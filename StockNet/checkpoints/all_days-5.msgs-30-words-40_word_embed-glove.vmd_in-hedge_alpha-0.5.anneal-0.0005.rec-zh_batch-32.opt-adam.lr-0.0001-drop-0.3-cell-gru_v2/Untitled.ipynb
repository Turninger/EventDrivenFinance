```python
import re
import matplotlib.pyplot as plt

#读取train.log文件
with open('train-2023-12-10.log', 'r') as f:
    log_content = f.read()
with open('train.log.2023-12-14', 'r') as f:
    log_content += f.read()
with open('train.log.2023-12-17', 'r') as f:
    log_content += f.read()

with open('train.log', 'r') as f:
    log_content += f.read()
# 使用正则表达式匹配train_loss和train_acc的值
loss_pattern = r'train_loss = ([\d.]+)'
acc_pattern = r'train_acc = ([\d.]+)'

# 提取train_loss和train_acc的值
loss_matches = re.findall(loss_pattern, log_content)
acc_matches = re.findall(acc_pattern, log_content)

# 转换为浮点数列表
loss_values = [float(value) for value in loss_matches]
acc_values = [float(value) for value in acc_matches]

# 创建epoch列表
epochs = list(range(1, len(loss_values) + 1))

# 创建train_loss图表
plt.subplot(2, 1, 1)  # 创建第一个子图
plt.plot(epochs, loss_values)
plt.title('Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')

# 创建train_acc图表
plt.subplot(2, 1, 2)  # 创建第二个子图
plt.plot(epochs, acc_values)
plt.title('Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')

# 调整子图间的间距
plt.tight_layout()

# 显示图表
plt.show()
```
